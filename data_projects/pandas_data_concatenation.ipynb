```python
# para colocar a biblioteca em uso, devemos importar ela
import pandas as pd
```

```python
# criar o esqueleto do arquivo
# usaremos dicionarios para criar os dados
# dicionário funciona com a ideia de chave-valor ou chave-valores
```

```python
dados = {
    'nome': ['André', 'Paulo', 'Aline', 'Paola'],
    'idade': [36,25,30,18],
    'cidade': ['Sorocaba','Indaiatuba','São Paulo','Registro']
}

# A biblioteca PANDAS pode transformar uma biblioteca em uma "tabela"
# Esse pedaço é chamado de DataFrame
#Ex.:
df_dados = pd.DataFrame(dados)

display(df_dados)
```

```python
# criar arquivo csv

# para criar, precisamos de dados
#chamamos os dados e atribuimos a função "to_"

df_dados.to_csv("teste_pandas.csv", index=False)
```

```python
# como ler um arquivo que me enviaram
# criar variavel e aplicar o comando read

# df_ler_dados = pd.read_csv('caminho_do_arquivo')
df_ler_dados = pd.read_csv('/content/teste_pandas.csv')
```

```python
# vamos popular a tabela

novos_dados = {
    'nome':['Flavia','Cláudio','Fernanda'],
    'idade':[50,63,32],
    'cidade':['Sorocaba','Iguape','Curitiba']
}
df_novos_dados = pd.DataFrame(novos_dados)
# Criar variavel para receber os dados, chama a PANDAS com a função DataFrame
#juntar = concatenar o antigo com o novo
df_dados_juntos = pd.concat([df_ler_dados,df_novos_dados], ignore_index=True)

#voltar para o CSV
df_dados_juntos.to_csv('teste_dados.csv',index=False)
display(df_dados_juntos)
```
